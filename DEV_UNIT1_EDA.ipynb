```python
#1. Let's assume you have five courses in your academic year:
subjects = ["Python programming", "Numerical methods", "Data Science", "DBMS", "DAA"]
# And you planned to obtain the following grades in each subject:
plannedGrade = [90, 95, 92, 68, 68, 90]
# However, after your final examination, these are the grades you got:
actualGrade = [75, 89, 89, 80, 80, 75]
import numpy as np
import matplotlib.pyplot as plt
#Prepare the dataset and set up theta:
theta = np.linspace(0, 2 * np.pi, len(plannedGrade))
# Initialize the plot with the figure size and polar projection:
plt.figure(figsize = (10,6))
plt.subplot(polar=True)

# Get the grid lines to align with each of the subject names:
(lines,labels) = plt.thetagrids(range(0,360,int(360/len(subjects))), (subjects))

# Use the plt.plot method to plot the graph and fill the area under it:
plt.plot(theta, plannedGrade)
plt.fill(theta, plannedGrade, 'b', alpha=0.2)

# Now, we plot the actual grades obtained:
plt.plot(theta, actualGrade)

# We add a legend and a nice comprehensible title to the plot:
plt.legend(labels=('Planned Grades','Actual Grades'),loc=1)
plt.title("Plan vs Actual grades by Subject")


plt.show()

```

```python
# histogram


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/drive/My Drive/DEA/ds_salaries.csv')

sal=data.salary

# Plot the distribution of sarary in USD:
nbins = 300
n, bins, patches = plt.hist(sal, bins=nbins)


# Add labels to the axes and a title:
plt.xlabel("salary  ")
plt.ylabel("Frequency")
plt.title("Distribution of salary  ")
```

```python
# histogram
plt.figure(figsize = (10,6))
nbins = 20
salusd = data.salary_in_usd
n, bins, patches = plt.hist(salusd, bins=nbins,density=1)
plt.xlabel("Salary in usd")
plt.ylabel("Frequency")
plt.title("Distribution of salary in usd ")
#Draw a green vertical line in the graph at the average salary in usd:
plt.axvline(x=salusd.mean(), linewidth=3, color = 'g')
```

```python
# histogram
plt.figure(figsize = (10,6))
nbins = 20
salusd = data.salary_in_usd
n, bins, patches = plt.hist(salusd, bins=nbins,density=1)
plt.xlabel("Salary in usd")
plt.ylabel("Frequency")
plt.title("Distribution of salary in usd ")
#Draw a green vertical line in the graph at the average salary in usd:
plt.axvline(x=salusd.mean(), linewidth=3, color = 'g')

#Compute the mean and standard deviation of the dataset:
mu = data.salary_in_usd.mean()
sigma = data.salary_in_usd.std()

#Add a best-fit line for the normal distribution:
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma* (bins - mu))**2))
#Plot the normal distribution:
plt.plot(bins, y, '--')
# Display the plot:
plt.show()
```

```python
#Should I work Part Time,Full Time, Contract base or as a Freelancer to get higher salary?
# Which  Employees are Earning higher salary ?

d=data.groupby(['employment_type'])['salary_in_usd'].mean()
ax=sns.barplot(d.index,d)
ax.set_xticklabels(['Contract','Freelance','Full time','Part time'])
ax.set_title('Average salary_in_usd per employment_type')
ax.set_ylabel('Average salary_in_usd')
```

```python
#Observation: Contract base Employees are Earning higher salary
```

```python
#What are the salaries of Data Science jobs based on experience level?
# which  Level job have high salary

d=data.groupby(['experience_level'])['salary_in_usd'].mean()
ax=sns.barplot(d.index,d)
ax.set_xticklabels(['Entry Level', 'Expert', 'Mid-Level', 'Senior Engineer'])
ax.set_title('Average salary_in_usd per Experience level')
ax.set_ylabel('Average salary_in_usd')
```

```python
# What is Stackplot ?
# It is used to draw a stacked area plot. 
#It shows each part stacked onto one another
#It displays the complete data for visualization. 



# House loan Mortgage cost per month for a year
houseLoanMortgage = [9000, 9000, 8000, 9000, 8000, 9000, 9000, 9000, 9000, 8000, 9000, 9000]

# Utilities Bills for a year
utilitiesBills = [4218, 4218, 4218, 4218, 4218, 4218, 4219, 2218, 3218, 4233, 3000, 3000]

# Transportation bill for a year
transportation = [782, 900, 732, 892, 334, 222, 300, 800, 900, 582, 596, 222]

# Car mortgage cost for one year
carMortgage = [700, 701, 702, 703, 704, 705, 706, 707, 708, 709, 710, 711]

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

months= [x for x in range(1,13)]

# Create placeholders for plot and add required color
plt.plot([],[], color='sandybrown', label='houseLoanMortgage')
plt.plot([],[], color='tan', label='utilitiesBills')
plt.plot([],[], color='bisque', label='transportation')
plt.plot([],[], color='darkcyan', label='carMortgage')

# Add stacks to the plot
plt.stackplot(months, houseLoanMortgage, utilitiesBills, transportation,carMortgage, colors=['sandybrown', 'tan', 'bisque', 'darkcyan'])
plt.legend()

# Add Labels
plt.title('Household Expenses')
plt.xlabel('Months of the year')
plt.ylabel('Cost')
# Display on the screen
plt.show()
```
